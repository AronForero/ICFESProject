```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
vs_LR_2000 = pd.read_csv('Final_Results/Final_LR_vs.csv', sep=';')
vs_DT_2000 = pd.read_csv('Final_Results/Final_DT_vs.csv', sep=';')
vs_RF_2000 = pd.read_csv('Final_Results/Final_RF_vs.csv', sep=';')

vs_LR_2014 = pd.read_csv('Final_Results/Final_LR_vs_2014.csv', sep=';')
vs_DT_2014 = pd.read_csv('Final_Results/Final_DT_vs_2014.csv', sep=';')
vs_RF_2014 = pd.read_csv('Final_Results/Final_RF_vs_2014.csv', sep=';')
```

```python
def fix_index(df):
    df.index = df['Unnamed: 0'].values
    df = df.drop('Unnamed: 0', axis=1)
    return(df)
```

```python
vs_LR_2000 = fix_index(vs_LR_2000)
vs_DT_2000 = fix_index(vs_DT_2000)
vs_RF_2000 = fix_index(vs_RF_2000)

vs_LR_2014 = fix_index(vs_LR_2014)
vs_DT_2014 = fix_index(vs_DT_2014)
vs_RF_2014 = fix_index(vs_RF_2014)
```

Plotting the DataFrames

```python
vs_LR_2000.plot(subplots=True, layout=(5, 3), title='Prediction vs Period', figsize=(16,15), ylim=(-1.2, 100.2));
plt.savefig('Final_Results/plot_target_LR_scores_vs_final')
```

```python
vs_DT_2000.plot(subplots=True, layout=(5, 3), title='Prediction vs Period', figsize=(16,15), ylim=(-1.2));
plt.savefig('Final_Results/plot_target_DT_scores_vs_final')
```

```python
vs_RF_2000.plot(subplots=True, layout=(5, 3), title='Prediction vs Period', figsize=(16,15), ylim=(-1.2));
plt.savefig('Final_Results/plot_target_RF_scores_vs_final')
```
