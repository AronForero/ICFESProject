New function made, to make the work easier.  

Returns the relation

```python
def get_relation(df1, df2, c):
    Geo_N = pd.DataFrame(df1[c].value_counts()).sort_values(by=c)
    Geo_C = pd.DataFrame(df2[c].value_counts()).sort_values(by=c)
    relation = pd.DataFrame({'col1':Geo_C.index, 'col2': Geo_N.index})
    return(relation)
```

```python
Geo_relation_2000_1 = get_relation(O_data_2000_1, data_2000_1, 'ESTU_RESIDE_DEPTO')
```
