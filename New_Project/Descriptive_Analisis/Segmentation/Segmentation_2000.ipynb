```python
fig, axes = plt.subplots(7, 4, sharex=False, figsize=(13, 17))
axs = [i for i in axes.flatten()]
for n, i in enumerate(Geo_Groups.groups.keys()):
    dic = Geo_Groups.get_group(i)
    ax = axs[n]
    ax.plot(dic['PUNT_MATEMATICAS'].sort_values().values)
    ax.set_title('PUNT_BIOLOGIA en la region '+'\n'+str(dic['COLE_DEPTO_UBICACION'].unique())+'.')
    ax.grid()
plt.tight_layout(h_pad=2)
```
