#### Now i have to plot the columns individualy, to check if there is some relation between the columns with the Biology scores

```python
data_2000_1 = data_2000_1.sort_values(by='PUNT_BIOLOGIA')
```

```python
y_list = ['PUNT_BIOLOGIA', 'PUNT_MATEMATICAS', 'PUNT_FILOSOFIA', 'PUNT_FISICA', 'PUNT_HISTORIA', 'PUNT_QUIMICA', 
          'PUNT_LENGUAJE', 'PUNT_GEOGRAFIA', 'PUNT_INTERDISCIPLINAR', 'PUNT_IDIOMA']
```

_______

Cruzo las tablas para experimentar, y ver la relacion entre el idioma escogido y el puntaje

```python
Id_PUNT = pd.crosstab(data_2000_1.COD_IDIOMA, data_2000_1.PUNT_IDIOMA)
```

```python
Id_PUNT.head()
```

```python
fig, axes = plt.subplots(1, 3, sharex=False, figsize=(13, 4))
axs = [i for i in axes.flatten()]
for n, i in enumerate(Id_PUNT.index):
    dic = Id_PUNT.loc[i]
    ax = axs[n]
    ax.scatter(range(len(dic)), dic)
    ax.set_title('Comportamiento de la columna: '+'\n'+str(i)+'.')
    ax.grid()
plt.tight_layout(h_pad=2)
```

______

```python
Grouped = data_2000_1.groupby(['COLE_DEPTO_UBICACION'])[y_list].agg('mean')
```

```python
Grouped.head()
```

```python
fig, axes = plt.subplots(4, 3, sharex=False, figsize=(15, 17))
axs = [i for i in axes.flatten()]
for n, i in enumerate(Grouped.columns):
    dic = Grouped[i]
    ax = axs[n]
    ax.bar(range(len(dic.index)), dic.values, alpha=0.5)
    ax.set_title('PUNTAJE en: '+str(i)+'.')
    ax.set_xticks(range(len(dic.index)))
    ax.set_xticklabels(dic.index, rotation=90)
    ax.grid()
plt.tight_layout(h_pad=2)
```
