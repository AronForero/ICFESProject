Hago la union de las columnas, y guardo el resultado en un conjunto llamado ***U***

```python
U = set()
U = seto[0]
for i in seto:
    U = U.union(i)
```

```python
#longitud del conjunto
len(U)
```

Creo una lista de listas llamada ***datac***, en donde se pondran listas que contendran 1s y 0s indicando si el campo esta(1) o no esta(0) en el fichero correspondiente a la columna

```python
datac = []
for i in U:
    temp = []
    temp.append(i)
    for j in seto:
        if i in j:
            temp.append(1)
        else:
            temp.append(0)
    datac.append(temp)
```

Se hace un sort para que esta lista de listas quede organizada en orden alfabetico

```python
datac.sort()
```

Elimino el primer campo, que corresponde al nombre del campo que se evaluo, y simplemente fue util para el ordenamiento anterior

```python
for i in datac:
    i.pop(0)
```
