# Analysis of the Individual Predictions 2000 and 2014

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Individual_LR_2000 = pd.read_csv('Final_Results/Final_LR_individual.csv', sep=';')
Individual_DT_2000 = pd.read_csv('Final_Results/Final_DT_individual.csv', sep=';')
Individual_RF_2000 = pd.read_csv('Final_Results/Final_RF_individual.csv', sep=';')

Individual_LR_2014 = pd.read_csv('Final_Results/Final_LR_2014_individual.csv', sep=';')
Individual_DT_2014 = pd.read_csv('Final_Results/Final_DT_2014_individual.csv', sep=';')
Individual_RF_2014 = pd.read_csv('Final_Results/Final_RF_2014_individual.csv', sep=';')
```

```python
def fix_index(df):
    df.index = df['Unnamed: 0'].values
    df = df.drop('Unnamed: 0', axis=1)
    return(df)
```

```python
Individual_LR_2000 = fix_index(Individual_LR_2000)
Individual_DT_2000 = fix_index(Individual_DT_2000)
Individual_RF_2000 = fix_index(Individual_RF_2000)

Individual_LR_2014 = fix_index(Individual_LR_2014)
Individual_DT_2014 = fix_index(Individual_DT_2014)
Individual_RF_2014 = fix_index(Individual_RF_2014)
```

Graficando los DataFrames

```python
Individual_LR_2000.plot(subplots=True, layout=(5, 3), title='Prediction vs Period', figsize=(16,15));
plt.savefig('Final_Results/plot_target_scores_final')
```

```python
Individual_DT_2000.plot(subplots=True, layout=(5, 3), title='Prediction vs Period', figsize=(16,15));
plt.savefig('Final_Results/plot_DT_target_scores_final')
```

```python
Individual_RF_2000.plot(subplots=True, layout=(5, 3), title='Prediction vs Period', figsize=(16,15));
plt.savefig('Final_Results/plot_RF_target_scores_final')
```
