Quito el signo de Porcentaje (%) para poder trabajar con todo numerico

Grafico la linea que describen las tasas mostradas en el Dataframe "D_Anual"

```python
plt.figure(figsize=(10,8))

plt.subplot(211)
plt.title("Tasa de Desercion semestral")
plt.plot(np.linspace(2001,2016, 32), D_Anual.iloc[0][1:])
#plt.xticks(np.linspace(2001,2016, 32), L, rotation="vertical");

plt.subplot(212)
plt.title("Tasa de Retencion Semestral")
plt.plot(np.linspace(2001,2016, 32), D_Anual.iloc[1][1:])
#plt.xticks(np.linspace(2001,2016, 32), L, rotation="vertical");
```

Creo una matriz de correlacion para mostrar un Heatmap

```python
Cor_D_Anual = D_Anual.corr()
```

Heatmaps con Pyplot y con Seaborn... Se obtienen diferentes mapas de calor

```python
plt.figure(figsize=(8,6))
plt.subplot(221)
plt.imshow(Cor_D_Anual, cmap="hot", interpolation="nearest")
plt.subplot(222)
sns.heatmap(Cor_D_Anual, cmap="inferno")
```
