   ### ERROR de condificiacion

Mismo Error de Codificacion encontrado en anteriores archivos  
Mediante la herramienta "file" de Linux se pudo observar que **TODOS, Excepto 2 (UTF-8) ** archivos tienen una codificacion "ISO-8859"  

Se aplicara una conversion de codificacion con la herramienta "iconv" de linux a cada archivo, con el fin de dejar todos los archivos en codificacion **UTF-8** asi:

<i>!iconv -f [ISO-8859-1] -t [utf-8] RutaDelArchivo.txt > RutaDelArchivo.NuevoNombredelArchivo.txt</i>

Una vez realizada la conversion, se eliminara el archivo que esta en la codificacion vieja asi:  

<i>!rm RutadelArchivoViejo.txt</i>

```python
!iconv -f [ISO-8859-1] -t [utf-8] ../DatosFTPICFES/SABER11/SB11-CLASIFICACION-PLANTELES/establecimientos_20162.txt > ../DatosFTPICFES/SABER11/SB11-CLASIFICACION-PLANTELES/establecimientos_20162-utf8.txt 
!rm ../DatosFTPICFES/SABER11/SB11-CLASIFICACION-PLANTELES/establecimientos_20162.txt
```
