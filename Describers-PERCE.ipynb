### Functions 

I create a list called N to have a numerical way of the letters, taking into account that the new characters doesn't be repeated in any other column of the table.

Then...

I made a function that change the string type to an integer type automatically

```python
def letters_to_num(obj):
    L = ["N", "S", "A", "B", "C", "D", "E", "F", "G", "H"]
    N = [-1, -2, -10, -11, -12, -13, -14, -15, -16, -17]
    x = obj.replace(L, N)
    return x
```

Then, i create a function that tell me how many columns are in the describe() of the pandas object

```python
def shape_cols(obj):
    print("Columns found: ", obj.describe().shape[1])
```

#### Now, we have 52 columns

```python
PERCE_Cuestionario_Alumn1 = letters_to_num(PERCE_Cuestionario_Alumn)
```

```python
shape_cols(PERCE_Cuestionario_Alumn1)
```

```python
PERCE_Cuestionario_Alumn1.describe()
```
